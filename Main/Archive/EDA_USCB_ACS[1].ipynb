```python
import pandas as pd
import requests
```

```python
# Define the API endpoint and parameters
url = 'https://api.census.gov/data/2017/acs/acs1'
apikey_USCB = 'b83c9e8cbf71b387db3681c39e0d7ef719f6344f'
# Define the parameters for the request
params = {
    'get': 'NAME,B25001_001E',  # B25001_001E is total households
    'for': 'county:037',        # County code 037 for Los Angeles
    'in': 'state:06',           # State code 06 for California
    'key': apikey_USCB          # Replace with your Census API key
}

# Make the API request
response = requests.get(url, params=params)

# # Print status code and content to debug
# print(f"Status Code: {response.status_code}")
# print(f"Response Content: {response.text}")

# Check the status and get the data
if response.status_code == 200:
    data = response.json()
    for item in data:
        print(item)
else:
    print(f"Error: {response.status_code}")
```

```python
import requests

# Define the API 7endpoint and parameters
years = [2017,2018,2019, 2020, 2021, 2022, 2023]
acs_data = {}
for year in years:
    # year = str(year)
    url = f'https://api.census.gov/data/{year}/acs/acs1'
    print(url)
    apikey_USCB = 'b83c9e8cbf71b387db3681c39e0d7ef719f6344f'

    # Define the parameters for the request
    params = {
        'get': 'NAME,B19013_001E,B19301_001E,B23025_005E,B23025_003E,B19083_001E,B01003_001E,B01002_001E,B05002_013E,B25077_001E',
        'for': 'state:*',
        'key': apikey_USCB  
    }

    # Make the API request
    response = requests.get(url, params=params)

    # Check the status and get the data
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data[1:], columns=data[0])
        acs_data[str(year)] = df
        # print(df.head())
        # print(response.text)
    else:
        print(f"Error: {response.status_code}")
```
